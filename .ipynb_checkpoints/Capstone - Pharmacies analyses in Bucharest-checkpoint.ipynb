In this section I will define a new function to get coordinates based on an address using google api. After that 
I will get latitude and longitude of the Bucharest city center

```python
google_api_key=""

import requests

def get_coordinates(api_key, address, verbose=False):
    try:
        url = 'https://maps.googleapis.com/maps/api/geocode/json?key={}&address={}'.format(api_key, address)
        response = requests.get(url).json()
        if verbose:
            print('Google Maps API JSON result =>', response)
        results = response['results']
        geographical_data = results[0]['geometry']['location']
        lat = geographical_data['lat']
        lon = geographical_data['lng']
        return [lat, lon]
    except:
        return [None, None]
    
address = 'Bucharest, Romania'
bucharest_center = get_coordinates(google_api_key, address)
print('Coordinate of {}: {}'.format(address, bucharest_center))
```

In the next sesction we will define various function to transform from speherical coordinates to UTM coordinates 
and also to calculate distance between two points in both coordinate sistems.

```python
import pyproj
import math
from math import sin, cos, sqrt, atan2, radians

def lonlat_to_xy(lon, lat):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=35, datum='WGS84')
    xy = pyproj.transform(proj_latlon, proj_xy, lon, lat)
    return xy[0], xy[1]

def calc_xy_distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx*dx + dy*dy)

def xy_to_lonlat(x, y):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=35, datum='WGS84')
    lonlat = pyproj.transform(proj_xy, proj_latlon, x, y)
    return lonlat[0], lonlat[1]

def calc_ll_distance(la1, lo1, la2, lo2):
    R = 6373.0
    lat1 = radians(la1)
    lon1 = radians(lo1)
    lat2 = radians(la2)
    lon2 = radians(lo2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c *1000
    return distance

def get_address(api_key, latitude, longitude, verbose=False):
    try:
        url = 'https://maps.googleapis.com/maps/api/geocode/json?key={}&latlng={},{}'.format(api_key, latitude, longitude)
        response = requests.get(url).json()
        if verbose:
            print('Google Maps API JSON result =>', response)
        results = response['results']
        address = results[0]['formatted_address']
        return address
    except:
        return None    
```

We will generate all the neighbourhoods with a 500m radius and we will plote them with folium

```python
bucharest_center_x, bucharest_center_y = lonlat_to_xy(bucharest_center[1], bucharest_center[0])

k = math.sqrt(3) / 2 # Vertical offset for hexagonal grid cells
x_min = bucharest_center_x - 6000
x_step = 600
y_min = bucharest_center_y - 6000 - (int(21/k)*k*600 - 12000)/2
y_step = 600 * k 

latitudes = []
longitudes = []
distances_from_center = []
xs = []
ys = []
for i in range(0, int(21/k)):
    y = y_min + i * y_step
    x_offset = 300 if i%2==0 else 0
    for j in range(0, 21):
        x = x_min + j * x_step + x_offset
        distance_from_center = calc_xy_distance(bucharest_center_x, bucharest_center_y, x, y)
        if (distance_from_center <= 6001):
            lon, lat = xy_to_lonlat(x, y)
            latitudes.append(lat)
            longitudes.append(lon)
            distances_from_center.append(distance_from_center)
            xs.append(x)
            ys.append(y)

print(len(latitudes), 'candidate neighborhood centers generated.')
```

```python
import folium

map_bucharest = folium.Map(location=bucharest_center, zoom_start=13)
folium.Marker(bucharest_center).add_to(map_bucharest)
for lat, lon in zip(latitudes, longitudes):
    folium.Circle([lat, lon], radius=300, color='red', fill=False).add_to(map_bucharest)
map_bucharest
```

We will search each of those neighbourghoods for the pharmacies location using google api

```python
def get_places_near_location(lat, lon, category, api_key, radius=500):

    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius={}&type={}&key={}'.format(
        lat,lon,radius, category, api_key)
    results = requests.get(url).json()['results']
    
    places=[]
    
    for item in results:

        lat=item['geometry']['location']['lat']
        lng=item['geometry']['location']['lng']

        try:
            rating=item['rating']
        except:
            rating=0
    
        name=item['name']
        tip=item['types']

        places.append([name,rating,lat,lng])
        
    return places
```

```python
import pandas as pd

radius=500
category='pharmacy'
api_key=''

print('Getting all pharmacies:', end='')
for lat, lon in zip(latitudes, longitudes):
    
    places=get_places_near_location(lat,lon,category,api_key,radius)

    pl=pd.DataFrame(places)
    
    df.append(pl)
    
    print(' .', end='')
    
print('Done')
```

In the next section we will define the area where we will search for future places.

```python
bucharest_center_x, bucharest_center_y = lonlat_to_xy(bucharest_center[1], bucharest_center[0])

roi_x_min = bucharest_center_x - 500
roi_y_max = bucharest_center_y + 2500
roi_width = 5000
roi_height = 5000
roi_center_x = roi_x_min + 1500
roi_center_y = roi_y_max - 1500
roi_center_lon, roi_center_lat = xy_to_lonlat(roi_center_x, roi_center_y)
roi_center = [roi_center_lat, roi_center_lon]

map_bucharest = folium.Map(location=roi_center, zoom_start=12)
HeatMap(pharmas_latlons).add_to(map_bucharest)
folium.Marker(bucharest_center).add_to(map_bucharest)
folium.Circle(roi_center, radius=1500, color='white', fill=True, fill_opacity=0.4).add_to(map_bucharest)

map_bucharest
```

It has a 1500m radius.

```python
k = math.sqrt(3) / 2 # Vertical offset for hexagonal grid cells
x_step = 100
y_step = 100 * k 
roi_y_min = roi_center_y - 1500

roi_latitudes = []
roi_longitudes = []
roi_xs = []
roi_ys = []
for i in range(0, int(51/k)):
    y = roi_y_min + i * y_step
    x_offset = 50 if i%2==0 else 0
    for j in range(0, 51):
        x = roi_x_min + j * x_step + x_offset
        d = calc_xy_distance(roi_center_x, roi_center_y, x, y)
        if (d <= 1501):
            lon, lat = xy_to_lonlat(x, y)
            roi_latitudes.append(lat)
            roi_longitudes.append(lon)
            roi_xs.append(x)
            roi_ys.append(y)

print(len(roi_latitudes), 'candidate neighborhood centers generated.')
```

We will search for each of those 821 centers if there are any pharmacies on a 500m radius area

```python
def count_pharma_nearby(x, y, pharma, radius=500):    
    count = 0
    for i in range(0,pharma.shape[0]-1):
        res_x = pharma.iloc[i,2]; res_y = pharma.iloc[i,3]
        d = calc_ll_distance(x, y, res_x, res_y)
        if d<=radius:
            count += 1
    return count

def find_nearest_pharma(x, y, pharma):
    d_min = 100000
    for i in range(0,pharma.shape[0]-1):
        res_x = pharma.iloc[i,2]; res_y = pharma.iloc[i,3]
        d = calc_ll_distance(x, y, res_x, res_y)
        if d<=d_min:
            d_min = d
    return d_min

roi_pharma_counts = []
roi_pharma_distances = []

print('Generating data on location candidates... ', end='')
for x, y in zip(roi_latitudes, roi_longitudes):
    count = count_pharma_nearby(x, y, df, radius=500)
    roi_pharma_counts.append(count)
    distance = find_nearest_pharma(x, y, df)
    roi_pharma_distances.append(distance)
print('done.')
```

We will define a new data fram which we will include two new columns. We want to know the number of sorounding pharmacies and also the minimum distance to one pharmacy

```python
df_roi_locations = pd.DataFrame({'Latitude':roi_latitudes,
                                 'Longitude':roi_longitudes,
                                 'X':roi_xs,
                                 'Y':roi_ys,
                                 'Pharmacies nearby':roi_pharma_counts,
                                 'Distance to a Pharmacy':roi_pharma_distances})
```

```python
import numpy as np

good_ph_count = np.array((df_roi_locations['Pharmacies nearby']<=2))
print('Locations with no more than two Pharmacies nearby:', good_ph_count.sum())

good_ph_distance = np.array(df_roi_locations['Distance to a Pharmacy']>=500)
print('Locations with no Pharmacy within 500m:', good_ph_distance.sum())

good_locations = np.logical_and(good_ph_count, good_ph_distance)
print('Locations with both conditions met:', good_locations.sum())

df_good_locations = df_roi_locations[good_locations]
```

We will use again KMeans to identify the clusters. Now we will use directly number of cluster to be 2. It looks preaty obvious

```python
from sklearn.cluster import KMeans

number_of_clusters = 2

good_xys = df_good_locations[['X', 'Y']].values
kmeans = KMeans(n_clusters=number_of_clusters, random_state=0).fit(good_xys)

cluster_centers = [xy_to_lonlat(cc[0], cc[1]) for cc in kmeans.cluster_centers_]

map_bucharest = folium.Map(location=roi_center, zoom_start=14)
folium.TileLayer('cartodbpositron').add_to(map_bucharest)
HeatMap(pharmas_latlons).add_to(map_bucharest)
folium.Circle(roi_center, radius=2500, color='white', fill=True, fill_opacity=0.4).add_to(map_bucharest)
folium.Marker(bucharest_center).add_to(map_bucharest)
for lon, lat in cluster_centers:
    folium.Circle([lat, lon], radius=500, color='red', fill=True, fill_opacity=0.25).add_to(map_bucharest) 
for lat, lon in zip(good_latitudes, good_longitudes):
    folium.CircleMarker([lat, lon], radius=2, color='blue', fill=True, fill_color='blue', fill_opacity=1).add_to(map_bucharest)
map_bucharest
```
